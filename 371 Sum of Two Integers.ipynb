Given two integers a and b, return the sum of the two integers without using the operators + and -.

 

Example 1:<br>
Input: a = 1, b = 2<br>
Output: 3


Example 2:<br>
Input: a = 2, b = 3<br>
Output: 5
 

Constraints:<br>
-1000 <= a, b <= 1000

```python
def getSum(a: int, b: int) -> int:

    while (b != 0):
        temp = (a & b) << 1
        a = a ^ b
        b = temp

    return a
```

```python
getSum(1,2)
```

```python
getSum(2,3)
```

```python
def getSum2(a: int, b: int) -> int:
    while b > 0:
        a, b = a ^ b, (a & b) << 1
    return a
```

```python
getSum2(1,2)
```

```python
getSum2(2,3)
```

```python
def getSum3(a: int, b: int) -> int:
    return math.log(math.exp(a)*math.exp(b))
```

```python
getSum3(1,2)
```

```python
getSum3(2,3)
```

```python
def getSum4(a: int, b: int) -> int:
    mask = 0xffffffff
    while (mask&b) > 0:
        a, b = a^b, (a&b)<<1
    return (mask&a) if b > 0 else a 
```

```python
getSum4(-2, 3)
```
