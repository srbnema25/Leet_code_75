```python
from collections import defaultdict
```

```python

def groupAnagrams(strs):

    res = defaultdict(list) 

    for s in strs:
        count = [0] * 26

        for c in s:
            count[ord(c) - ord("a")] += 1

        res[tuple(count)].append(s)

    return list(res.values())
```

```python
groupAnagrams(["eat","tea","tan","ate","nat","bat"])
```

```python
groupAnagrams([""])
```

```python
groupAnagrams(["a"])
```
