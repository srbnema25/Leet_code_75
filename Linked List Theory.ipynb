Python list is a dynamic array 

In a linked list if you wan to insert element at the beginning = O(1)<br>
Delete Element at beginning = O(1)<br>
Insert / Delete Element at the end = O(n)<br>

Linked List has two main benefits over an array :
1) You dont need to pre-allocate space <br>
2) Insertion is easier

Linked List Traversal = O(n) <br>
Accessing Element by Value = O(n)

In a Double Linked List, a node has an address to previous and next element both.

Array Vs Linked List

Indexing  ---  O(1) Vs O(n) <br>
Insert / Delete Element at Start --- O(n) Vs O(1) <br>
Insert / Delete Element at End --- O(1) - amortized Vs O(n) <br>
Insert Element in Middle --- O(n) Vs O(n)

```python
class Node:
    def __init__ (self, data = None, next_add = None):
        self.data = data
        self.next = next_add
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert_at_begining(self, data):
        node = Node(data, self.head)
        self.head = node
        
    def print(self):
        if self.head is None:
            print("Linked list is empty")
            return
        
        itr = self.head
        llstr = ""
        
        while itr:
            llstr += str(itr.data) + '-->'
            itr = itr.next
        
        print(llstr)
        
    def insert_at_end(self, data):
        if self.head is None:
            self.head = Node(data, None)
            return
        
        itr = self.head
        while itr.next:
            itr = itr.next
            
        itr.next = Node(data, None)
        
    def insert_values(self, data_list):
        
        self.head = None
        for data in data_list:
            self.insert_at_end(data)
            
    
    def get_length(self):
        count = 0
        
        itr = self.head
        while itr:
            count += 1
            itr = itr.next
            
        return count
    
    def remove_at_index(self, index):
        if index < 0 or index>= self.get_length():
            raise Exception("Invalid Index")
            
        if index == 0:
            self.head = self.head.next
            return
        
        count = 0 
        itr = self.head
        while itr:
            if count == index -1:
                itr.next = itr.next.next
                break
                
            itr = itr.next
            count += 1
    
    
    def insert_at_index(self, index, data):
        if index < 0 or index > self.get_length():
            raise Exception("Invalid Index")
            
        if index == 0:
            self.insert_at_begining(data)
            return
        
        count=0 
        itr = self.head
        temp = 0
        while itr:
            if count == index -1:
                itr.next = Node(data, itr.next)
                break
            
            itr = itr.next
            count += 1
```

```python
ll = LinkedList()
ll.insert_at_begining(5)
ll.insert_at_begining(89)
ll.insert_at_begining(43)
ll.print()
```

```python
ll.insert_at_end(777)
ll.print()
```

```python
ll = LinkedList()
ll.insert_values(["banana", "mango", "grapes", 'orange'])
ll.print()
```
