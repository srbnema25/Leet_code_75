# 3. Longest Substring Without Repeating Characters

Given a string s, find the length of the longest substring without duplicate characters.

 

Example 1:

Input: s = "abcabcbb"
Output: 3
Explanation: The answer is "abc", with the length of 3. Note that "bca" and "cab" are also correct answers.

Example 2:

Input: s = "bbbbb"
Output: 1
Explanation: The answer is "b", with the length of 1.

Example 3:

Input: s = "pwwkew"
Output: 3
Explanation: The answer is "wke", with the length of 3.
Notice that the answer must be a substring, "pwke" is a subsequence and not a substring.
 

Constraints:

0 <= s.length <= 5 * 104
s consists of English letters, digits, symbols and spaces.

# Time complexity = O(N3)

```python
def lengthOfLongestSubstring(s):
    longest_len = 0
    longest = ''
    for i in range(len(s)):
        for q in range(i+1, len(s)+1):
            if len(s[i:q])==len(set(s[i:q])):
                longest_len = max(longest_len, len(s[i:q]))
        
    return longest_len
```

Key operations

Each iteration does:

s[l:r]<br>
set(s[l:r])<br>
len(s[l:r])<br>

Cost of each
s[l:r] → O(k) where k = r-l<br>
set(s[l:r]) → O(k)<br>
len(s[l:r]) → O(1) but substring creation already O(k)<br>

So each iteration costs O(k).

Worst case:

s = "abcdefg...."

r grows up to n, and substring sizes grow gradually.

Total cost roughly:

1 + 2 + 3 + ... + n  ≈ O(n²)

Time Complexity : O(n²)
Space Complexity<br>
Temporary set and substring:<br>
O(n)

Other Issues<br>
This line:<br>
while (r <= len(s)) & (l <= len(s)):<br>
uses bitwise & instead of and, which is slower and less idiomatic.

**The Set Trap:** In 2nd solution, set(s[l:r]) creates a brand new set and re-hashes every character in the current window every single time the loop runs. If your window is 100 characters long, you do 100 operations just to check for duplicates.

```python
def lengthOfLongestSubstring(s: str) -> int:
    l, r = 0, 1
    max_string = 0

    while (r <= len(s)) & (l <= len(s)):
        if len(s[l:r]) == len(set(s[l:r])):
            max_string = max(len(s[l:r]), max_string)
            r += 1
        else:
            l += 1

    return max_string
```

True Sliding Window: Solution 3 maintains a single set. When the right pointer moves, it just adds one character. When a duplicate is found, it just removes characters from the left until the duplicate is gone. It never "re-reads" the middle of the window.

This uses the Sliding Window technique.

Key idea

Maintain a window [l,r] with unique characters.

When duplicate appears:

Move l until duplicate removed.

Important observation

Each character is:

added to set once

removed from set once

So total operations:

add ≤ n
remove ≤ n
Time Complexity
O(n)
Space Complexity

Set stores unique chars in window.

Worst case:

O(min(n, charset_size))

For ASCII → O(128) ≈ constant.

```python
def lengthOfLongestSubstring(s):
    charSet = set()
    l = 0
    res = 0
    
    for r in range(len(s)):
        while s[r] in charSet:
            charSet.remove(s[l])
            l += 1
            
        charSet.add(s[r])
        
        res = max(res, r-l+1)
        
    return res
```

Why this is faster

Instead of removing characters one-by-one:

while s[r] in charSet:<br>
    remove...

We jump l directly.

Example:

abba

Instead of:

remove a<br>
remove b

We directly move:<br>
l = last_seen['b'] + 1<br>


Complexity

Time: O(n)

Space: O(min(n, charset))

But fewer operations than Solution 2.

```python
def lengthOfLongestSubstring(s):
    last_seen = {}
    l = 0
    res = 0

    for r, c in enumerate(s):
        if c in last_seen:
            l = max(l, last_seen[c] + 1)

        last_seen[c] = r
        res = max(res, r - l + 1)

    return res
```
