Given the head of a linked list, remove the nth node from the end of the list and return its head.

Example 1:<br>
Input: head = [1,2,3,4,5], n = 2<br>
Output: [1,2,3,5]

Example 2:<br>
Input: head = [1], n = 1<br>
Output: []
    
Example 3:<br>
Input: head = [1,2], n = 1<br>
Output: [1]
 
Constraints:<br>
The number of nodes in the list is sz.<br>
1 <= sz <= 30<br>
0 <= Node.val <= 100<br>
1 <= n <= sz<br>

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
lst = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
```

```python
tail = lst
while tail:
    print(tail.val)
    tail = tail.next
```

```python
def removeNthFromEnd(head, n):

    dummy = ListNode(0, head)
    left = dummy
    right = head

    while n > 0 and right:
        right = right.next
        n -= 1

    while right:
        left = left.next
        right = right.next

    left.next = left.next.next

    return dummy.next
```

```python
out_lst = removeNthFromEnd(lst, 2)
tail = out_lst
while tail:
    print(tail.val)
    tail = tail.next
```

```python
def removeNthFromEnd2(head, n):

    left = head
    right = head

    while n+1 > 0 and right:
        right = right.next
        n -= 1

    while right:
        left = left.next
        right = right.next

    left.next = left.next.next

    return head
```

```python
out_lst = removeNthFromEnd2(lst, 2)
tail = out_lst
while tail:
    print(tail.val)
    tail = tail.next
```
